```python
# label encoding
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
le = LabelEncoder()
```

```python
df['Churn'] = le.fit_transform(df['Churn']) #yes-1, no-0
```

```python
# one hot encoding for the other categorical data
df=pd.get_dummies(df, drop_first=True)
```

```python
df['Churn'].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X=df.drop('Churn', axis=1)
y=df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
acc=accuracy_score(y_test, y_pred)
```

```python
acc
```

```python
report=classification_report(y_test, y_pred)
```

```python
report
```

```python
matrix = confusion_matrix(y_test, y_pred)
```

```python
matrix
```

```python
from sklearn.model_selection import GridSearchCV

log_reg = LogisticRegression(max_iter=1000)
```

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10],              # Regularization strength
    'penalty': ['l1', 'l2'],              # L1/L2 regularization
    'solver': ['liblinear', 'saga'],      # Solvers that support l1 and l2
    'class_weight': [None, 'balanced']    # Handle imbalance
}
```

```python
# Step 2: GridSearchCV for Recall
grid_search = GridSearchCV(
    estimator=log_reg,
    param_grid=param_grid,
    scoring='recall',   # Focus on recall
    cv=5,
    n_jobs=-1,
    verbose=1
)
```

```python
grid_search.fit(X_train, y_train)

# Step 3: Best model
best_model = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)
```

```python
Y_pred = best_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, Y_pred))
print("Recall:", classification_report(y_test, Y_pred, digits=3))
```

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn.metrics import precision_score, recall_score

# --- Default Logistic Regression ---
default_model = LogisticRegression(max_iter=1000)
default_model.fit(X_train, y_train)
y_pred_default = default_model.predict(X_test)

default_results = {
    'Model': 'Default Logistic Regression',
    'Accuracy': accuracy_score(y_test, y_pred_default),
    'Recall (Churn)': recall_score(y_test, y_pred_default),
    'Precision (Churn)': precision_score(y_test, y_pred_default)
}

# --- Recall-Optimized Logistic Regression ---
y_pred_recall = best_model.predict(X_test)  # from hyper-tuned model

recall_results = {
    'Model': 'Recall-Optimized Logistic Regression',
    'Accuracy': accuracy_score(y_test, y_pred_recall),
    'Recall (Churn)': recall_score(y_test, y_pred_recall),
    'Precision (Churn)': precision_score(y_test, y_pred_recall)
}

# Combine into DataFrame
comparison_df = pd.DataFrame([default_results, recall_results])
comparison_df
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

# Probabilities from recall-optimized model
y_probs = best_model.predict_proba(X_test)[:, 1]

precision, recall, thresholds = precision_recall_curve(y_test, y_probs)
avg_precision = average_precision_score(y_test, y_probs)

plt.figure(figsize=(6,4))
plt.plot(recall, precision, label=f'Precision-Recall Curve (AP={avg_precision:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve (Recall-Optimized Model)')
plt.legend()
plt.grid(True)
plt.show()
```
